<h1>Missing Values Analysis</h1>

```python
missing_values = df.isnull().sum().sort_values(ascending=False)

missing_percent = (
    df.isnull().sum() / len(df) * 100
).sort_values(ascending=False)

missing_df = pd.DataFrame({
    "Missing Values": missing_values,
    "Missing Percentage": missing_percent
})

missing_df.head(20)
```

<h1>Drop High Missing Columns</h1>

```python
threshold = 0.7

missing_ratio = df.isnull().mean()

columns_to_drop = missing_ratio[
    missing_ratio > threshold
].index

df_clean = df.drop(columns=columns_to_drop)

print("Removed Columns:", len(columns_to_drop))
print("New Shape:", df_clean.shape)
```

<h1>Fill Missing Values</h1>

```python
df.isnull().sum()
```

```python
# Numeric columns
numeric_cols = df_clean.select_dtypes(
    include=np.number
).columns

df_clean[numeric_cols] = df_clean[numeric_cols].fillna(
    df_clean[numeric_cols].median()
)

# Categorical columns
categorical_cols = df_clean.select_dtypes(
    include="object"
).columns

for col in categorical_cols:
    df_clean[col] = df_clean[col].fillna(
        df_clean[col].mode()[0]
    )

df_clean.isnull().sum().sum()
```

<h1>Remove Duplicates</h>

```python
duplicates = df_clean.duplicated().sum()

print("Duplicate Rows:", duplicates)

df_clean = df_clean.drop_duplicates()

print(
    "Shape After Removing Duplicates:",
    df_clean.shape
)
```

<h1>Keep Countries Only</h1>

```python
df_countries = df_clean[
    df_clean["iso_code"].notnull()
].copy()

print(df_countries.shape)

df_countries.head()
```

<h1>Select Important Columns</h1>

```python
selected_columns = [
    "country",
    "year",
    "population",
    "gdp",
    "coal_consumption",
    "oil_consumption",
    "gas_consumption",
    "solar_consumption",
    "wind_consumption",
    "renewables_consumption",
    "fossil_fuel_consumption",
    "electricity_generation"
]

# Keep only columns that actually exist
existing_columns = [
    col for col in selected_columns
    if col in df_countries.columns
]

print("Existing Columns:")
print(existing_columns)

viz_df = df_countries[existing_columns].copy()

viz_df.head()
```

```python
print(df_countries.columns.tolist())
```

<h1>Create Decade Column</h1>

```python
viz_df["decade"] = (
    viz_df["year"] // 10
) * 10

viz_df.head()
```

<h1>Clean Country Names</h1>

```python
viz_df["country"] = (
    viz_df["country"]
    .str.strip()
    .str.title()
)

viz_df.head()
```

<h1>Outlier Detection</h1>

```python
numeric_viz = viz_df.select_dtypes(
    include=np.number
)

Q1 = numeric_viz.quantile(0.25)
Q3 = numeric_viz.quantile(0.75)

IQR = Q3 - Q1

outliers = (
    (
        numeric_viz < (Q1 - 1.5 * IQR)
    ) |
    (
        numeric_viz > (Q3 + 1.5 * IQR)
    )
).sum()

outliers.sort_values(
    ascending=False
).head(10)
```

<h1>Export Cleaned CSV</h1>

```python
viz_df.to_csv(
    "owid_energy_visualization_cleaned.csv",
    index=False
)

print("Cleaned dataset saved successfully.")
```
